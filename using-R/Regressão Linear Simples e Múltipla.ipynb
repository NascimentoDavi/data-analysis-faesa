#### Exercício Número 4

##### A)

```r
horas <- c(8, 8, 12, 12, 16, 16, 20, 20, 24, 24)
erros <- c(6, 8, 6, 10, 8, 14, 12, 14, 12, 16)

modelo <- lm(erros ~ horas)

plot(horas, erros, main="Relação entre Erros e Horas sem Dormir",
     xlab="Horas sem Dormir", ylab="Número de Erros", pch=19, col="blue")
abline(modelo, col="red", lwd=2)  # Adicionando a reta de regressão

# Adicionando as estimativas de erros para alguns valores de horas
novas_horas <- c(18)
novos_erros <- predict(modelo, data.frame(horas=novas_horas))
points(novas_horas, novos_erros, col="green", pch=19)  # Estimativa para 18 horas
```

##### B)

```r
horas <- c(8, 8, 12, 12, 16, 16, 20, 20, 24, 24)
erros <- c(6, 8, 6, 10, 8, 14, 12, 14, 12, 16)

modelo <- lm(erros ~ horas)

novas_horas <- c(2, 6, 18, 25, 30)

# Estimativa dos erros para esses valores
novos_erros <- predict(modelo, data.frame(horas=novas_horas))

# Exibir as estimativas
data.frame(Horas_Sem_Dormir = novas_horas, Estimativa_Erros = novos_erros)
```
