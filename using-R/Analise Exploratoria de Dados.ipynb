```r
meses <- c(8, 9, 4, 5, 3, 6, 8, 6, 6, 8, 5, 5, 6, 4, 4)
setor <- c("C", "C", "I", "I", "I", "C", "C", "I", "I", "C", "C", "I", "C", "I", "I")
tamanho <- c("G", "M", "G", "M", "M", "P", "G", "M", "P", "M", "P", "P", "M", "M", "G")
```

##### (a) Dividir as empresas em grupos e calcular média e mediana

```r
comercio <- meses[setor == "C"]
industria <- meses[setor == "I"]

media_comercio <- mean(comercio)
mediana_comercio <- median(comercio)

media_industria <- mean(industria)
mediana_industria <- median(industria)

cat("Comércio - Média:", media_comercio, "Mediana:", mediana_comercio, "\n")
cat("Indústria - Média:", media_industria, "Mediana:", mediana_industria, "\n")
```

##### (b) Calcular desvio padrão e variância

```r
variancia_comercio <- var(comercio)
desvio_padrao_comercio <- sd(comercio)

variancia_industria <- var(industria)
desvio_padrao_industria <- sd(industria)

cat("Comércio - Variância:", variancia_comercio, "Desvio Padrão:", desvio_padrao_comercio, "\n")
cat("Indústria - Variância:", variancia_industria, "Desvio Padrão:", desvio_padrao_industria, "\n")
```

##### (c) Calcular para tamanhos (P, M e G)

```r
pequeno <- meses[tamanho == "P"]
medio <- meses[tamanho == "M"]
grande <- meses[tamanho == "G"]

media_pequeno <- mean(pequeno)
mediana_pequeno <- median(pequeno)
variancia_pequeno <- var(pequeno)
desvio_padrao_pequeno <- sd(pequeno)

media_medio <- mean(medio)
mediana_medio <- median(medio)
variancia_medio <- var(medio)
desvio_padrao_medio <- sd(medio)

media_grande <- mean(grande)
mediana_grande <- median(grande)
variancia_grande <- var(grande)
desvio_padrao_grande <- sd(grande)

cat("Pequeno - Média:", media_pequeno, "Mediana:", mediana_pequeno, "Variância:", variancia_pequeno, "Desvio Padrão:", desvio_padrao_pequeno, "\n")
cat("Médio - Média:", media_medio, "Mediana:", mediana_medio, "Variância:", variancia_medio, "Desvio Padrão:", desvio_padrao_medio, "\n")
cat("Grande - Média:", media_grande, "Mediana:", mediana_grande, "Variância:", variancia_grande, "Desvio Padrão:", desvio_padrao_grande, "\n")
```

#### Exercício 2

##### Dados das idades e pesos

```r
idades <- c(40, 40, 40, 40, 40, 40, 
             45, 45, 45, 45, 45, 45, 
             50, 50, 50, 50, 50, 50, 
             55, 55, 55, 55, 55, 55, 
             60, 60, 60, 60, 60, 60, 
             65, 65, 65, 65, 65, 65)

pesos <- c(55, 50, 68, 65, 62, 62,
           58, 56, 62, 65, 63, 63,
           60, 74, 70, 78, 76, 76,
           77, 78, 70, 72, 80, 80,
           70, 76, 74, 83, 85, 85,
           65, 82, 72, 82, 80, 80)
```

#####  (a) Calcular média, mediana, desvio padrão e variância para cada grupo de idade

```r
resultados <- data.frame(Idade = c(40, 45, 50, 55, 60, 65))

for (idade in unique(idades)) {
  pesos_idade <- pesos[idades == idade]
  
  media <- mean(pesos_idade)
  mediana <- median(pesos_idade)
  variancia <- var(pesos_idade)
  desvio_padrao <- sd(pesos_idade)
  
  resultados[resultados$Idade == idade, "Média"] <- media
  resultados[resultados$Idade == idade, "Mediana"] <- mediana
  resultados[resultados$Idade == idade, "Variância"] <- variancia
  resultados[resultados$Idade == idade, "Desvio Padrão"] <- desvio_padrao
}

print(resultados)
```

##### (b) Análise do comportamento do peso com o aumento da idade

```r
plot(resultados$Idade, resultados$Média, type = "b", 
     xlab = "Idade (anos)", ylab = "Peso (Kg)", 
     main = "Média de Peso por Idade", col = "blue", pch = 19)
```

```r
# Dados das notas
notas <- data.frame(
  Seção = c(rep("Pessoal", 8), rep("Técnica", 7), rep("Vendas", 10)),
  Direito = c(9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9, 9),
  Política = c(9.0, 6.5, 9.0, 6.0, 6.5, 6.5, 9.0, 6.0, 9.0, 9.0, 7.0, 5.5, 6.0, 8.0, 7.0, 9.0, 6.0, 6.5, 6.0, 9.0, 6.5, 7.0, 6.0, 9.0, 7.0),
  Estatística = c(9, 9, 8, 8, 9, 10, 8, 8, 9, 8, 10, 7, 7, 9, 2, 8, 7, 7, 9, 8, 9, 7, 8, 9, 7)
)
```

##### (a) Cálculo das estatísticas descritivas

```r
calcular_estatisticas <- function(x) {
  media <- mean(x)
  mediana <- median(x)
  moda <- as.numeric(names(sort(table(x), decreasing = TRUE)[1])) 
  desvio_padrao <- sd(x)
  coef_variacao <- (desvio_padrao / media) * 100
  
  return(c(Media = media, Mediana = mediana, Moda = moda, Desvio_Padrao = desvio_padrao, Coeficiente_Variacao = coef_variacao))
}

estatisticas_direito <- calcular_estatisticas(notas$Direito)
estatisticas_politica <- calcular_estatisticas(notas$Política)
estatisticas_estatistica <- calcular_estatisticas(notas$Estatística)

cat("Estatísticas - Direito:\n")
print(estatisticas_direito)

cat("\nEstatísticas - Política:\n")
print(estatisticas_politica)

cat("\nEstatísticas - Estatística:\n")
print(estatisticas_estatistica)
```

##### (b) Boxplots para comparar as distribuições

```r
par(mfrow = c(1, 3)) 

boxplot(notas$Direito, main = "Boxplot - Direito", ylab = "Notas", col = "lightblue")
boxplot(notas$Política, main = "Boxplot - Política", ylab = "Notas", col = "lightgreen")
boxplot(notas$Estatística, main = "Boxplot - Estatística", ylab = "Notas", col = "lightcoral")
```

##### (c) Comparando o aproveitamento na disciplina Estatística segundo a seção

```r
boxplot(notas$Estatística ~ notas$Seção, main = "Aproveitamento em Estatística por Seção", 
        ylab = "Notas em Estatística", col = c("lightblue", "lightgreen", "lightcoral"))
```

#### Exercício 12

##### (a) Probabilidade de a medida exceder 13 mAh

```r
mu <- 10
sigma <- sqrt(4)  

X_a <- 13
Z_a <- (X_a - mu) / sigma
prob_a <- 1 - pnorm(Z_a) 
cat("Probabilidade de a medida exceder 13 mAh:", prob_a, "\n")
```

##### (b) Probabilidade de a medida da corrente estar entre 9 e 11 mAh

```r
X_b1 <- 9
X_b2 <- 11
Z_b1 <- (X_b1 - mu) / sigma  
Z_b2 <- (X_b2 - mu) / sigma  

prob_b <- pnorm(Z_b2) - pnorm(Z_b1)
cat("Probabilidade de a medida da corrente estar entre 9 e 11 mAh:", prob_b, "\n")
```
